# Introduction to Neural network Regression with TensorFlow

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## What is regression?
Regression is a type of supervised learning where the goal is to predict a continuous numerical value. For example, predicting the price of a house based on its size, number of bedrooms, etc.
## What is a neural network?
A neural network is a computational model inspired by the structure and function of the human brain. It consists of layers of interconnected nodes (neurons) that can learn to recognize patterns in data. Neural networks are particularly powerful for regression tasks because they can capture complex relationships between input features and the target variable.
## Building a neural network regression model with TensorFlow
To build a neural network regression model with TensorFlow, you can follow these steps: 
1. Import the necessary libraries (TensorFlow, NumPy, etc.).
2. Prepare your dataset (split into features and target variable, and then into training and testing
sets).
3. Define the architecture of your neural network (number of layers, number of neurons in each layer, activation functions, etc.).
4. Compile the model (specify the loss function, optimizer, and metrics).   
5. Train the model on the training data.
6. Evaluate the model on the testing data.
7. Make predictions with the trained model.

### Inputs and Outputs of a neural network regression modelMP 4
The inputs to a neural network regression model are typically the features of your dataset (e.g., size of the house, number of bedrooms, etc.). The output is a single continuous value that represents the predicted target variable (e.g., price of the house).

### Anatomy of a neural network regression model
A neural network regression model typically consists of the following components:   
- Input layer: This layer receives the input features of the dataset.
- Hidden layers: These layers perform computations on the input data and learn to capture complex relationships between the features and the target variable. The number of hidden layers and the number of neurons in each layer can be adjusted based on the complexity of the problem.
- Output layer: This layer produces the final prediction, which is a single continuous value in regression tasks.

# Hyperparameters
- Input layer Shape: (None, 1)  # Input shape for regression (e.g., one feature)
- Hidden layer(s) - Problem specific (e.g., 64 neurons, ReLU activation)
- Output layer Shape: (None, 1)  # Output shape for regression (e.g., one continuous value)
- Neurons per hidden layer - Problem specific (e.g., 64, 128, etc.)
- Activation function for hidden layers - Problem specific (e.g., ReLU, sigmoid, etc.)
- Output Activation function - Problem specific (e.g., linear for regression)
- loss function - Problem specific (e.g., mean squared error for regression)
- optimizer - Problem specific (e.g., Adam, SGD, etc.)

```python
# Example of model ( specific to regression )
modelex = tf.keras.Sequential([
    tf.keras.Input(shape=(3,)),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(75, activation='relu'),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')  # Linear activation for regression
])
# Compile the model with a regression-specific loss function and optimizer
modelex.compile(loss=tf.keras.losses.mae ,
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  metrics=["mae"])
```

#### Creating Data to view and fit

```python
 # create features
x=np.array([-7,-4,-1,2,5,8,11,14])
# create labels
y=np.array([-14,-8,-2,4,10,16,22,28])
# Visualize the data
plt.scatter(x,y);
plt.xlabel("Feature")
plt.ylabel("Label")
plt.show()
```

## Steps in modeling with TensorFlow
1. Create a demo tensor for our housing price prediction model
2. Create features and labels for our model
3. Creating a model - define the input, hidden and output layers
4. Compile the model - define the loss function ( in other word, the fuction which tells us how wrong our predictions are ) and optimizer ( tells the model how to improve the patterns it is learning ) and evaluation metrics ( tells us how well our model is performing )
5. Fit the model to the training data - let the model try to find patterns between features and labels
6. Evaluate the model on the test data - see how well it performs on data it has never seen before
7. Make predictions with the trained model - use the model to predict on new data ( make some house predictions )


```python
# set random seed
tf.random.set_seed(42)
# 1.create a model using the sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation=tf.keras.activations.relu), # hidden layer with 100 neurons and relu activation function
    tf.keras.layers.Dense(1) # output layer with 1 neuron ( for regression ) and default linear activation function 
])
# 2.Compile the model with a regression-specific loss function and optimizer 
model.compile(loss=tf.keras.losses.mae, # mae = mean absolute error
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01)) # Adam = adaptive moment estimation
```

```python
# 3. Fit the model to the training data - let the model try to find patterns between features and labels
history = model.fit(tf.expand_dims(x, axis=-1), y, epochs=100)
```

### Checking for Overfitting and Underfitting
To check if a model is overfitting or underfitting, we typically monitor its performance on both **training data** and a separate **validation dataset** over time (epochs).

- **Underfitting**: The model fails to learn the patterns, meaning both training loss and validation loss remain high. Possible fixes: Add more neurons/layers, train longer, use a different optimizer.
- **Overfitting**: The model memorizes the training data but fails to generalize. This happens when **training loss goes down** but **validation loss starts going up** (or plateaus higher than training loss). Possible fixes: Add more data, add regularization (like Dropout), simplify the model, or use Early Stopping.

Since our dataset here is very small, we will simply plot the training loss curve obtained from the `history` object returned by `model.fit()`. If we had a larger dataset, we would pass `validation_split=0.2` during `.fit()` and plot `history.history['val_loss']` alongside the training loss.

```python
# Let's plot our loss curves
import pandas as pd

# Convert history object into a pandas dataframe
pd.DataFrame(history.history).plot(figsize=(10, 7), xlabel="epochs")
plt.title("Model Training Loss Curve")
plt.show()
```

```python
# 4. Evaluate the model on the test data - see how well it performs on data it has never seen before
model.evaluate(tf.expand_dims(x, axis=-1), y)
# 5. Make predictions with the trained model - use the model to predict on new data ( make some house predictions )
y_pred = model.predict(tf.expand_dims(x, axis=-1))
y_pred
# accuracy of the model
plt.scatter(x,y, label="True values")
plt.scatter(x,y_pred, label="Predicted values")
plt.xlabel("Feature")
plt.ylabel("Label")
plt.legend()
plt.show()
# Accuracy as percentage
mae = tf.keras.losses.mae(y, tf.squeeze(y_pred))
mape = tf.keras.losses.mape(y, tf.squeeze(y_pred))
print(f"MAE: {tf.reduce_mean(mae).numpy():.2f}")
print(f"MAPE: {tf.reduce_mean(mape).numpy():.2f}%")

accuracy = 100 - tf.reduce_mean(mape).numpy()
print(f"Accuracy: {accuracy:.2f}%")
```
