## Shuffle , Reshape the order of elements in a tensor

```python
# shuffle a tensor
# we do this to make sure the order of our data is random and not in a specific order (e.g. sorted)
shuffled_tensor = tf.random.shuffle(random_tensor)
print("Shuffled tensor:\n", shuffled_tensor)
# reshape a tensor
reshaped_tensor = tf.reshape(random_tensor, shape=(4, 3))
print("Reshaped tensor:\n", reshaped_tensor)
# get the min and max values of a tensor
min_value = tf.reduce_min(random_tensor)
max_value = tf.reduce_max(random_tensor)
print("Min value:", min_value.numpy())
print("Max value:", max_value.numpy())
# get the mean and sum of a tensor
mean_value = tf.reduce_mean(random_tensor)
sum_value = tf.reduce_sum(random_tensor)
print("Mean value:", mean_value.numpy())
print("Sum value:", sum_value.numpy())
# get the index of max value
argmax_index = tf.argmax(random_tensor)         
print("Index of max value:", argmax_index.numpy())
```

```python
#indexing and slicing tensors and expanding tensors
tensor_2d = tf.constant([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
# indexing  
print("First row:", tensor_2d[0].numpy())
print("Second column:", tensor_2d[:, 1].numpy())
# slicing
print("Sliced tensor (rows 0-1, columns 1-2):\n", tensor_2d[0:2, 1:3].numpy())
# expanding dimensions  
expanded_tensor = tf.expand_dims(tensor_2d, axis=0)
print("Expanded tensor shape:", expanded_tensor.shape)  
# squeezing dimensions
squeezed_tensor = tf.squeeze(expanded_tensor)   
print("Squeezed tensor shape:", squeezed_tensor.shape)  
# stacking tensors
tensor_a = tf.constant([1, 2, 3])
tensor_b = tf.constant([4, 5, 6])
stacked_tensor = tf.stack([tensor_a, tensor_b], axis=0)
print("Stacked tensor:\n", stacked_tensor.numpy())
# unstacking tensors    
unstacked_tensors = tf.unstack(stacked_tensor, axis=0)
print("Unstacked tensors:") 
for t in unstacked_tensors:
    print(t.numpy())        
```

# Manipulating tensors (tensor operations)
**basic operations**
- Addition: `tf.add()`, `+`
- Subtraction: `tf.subtract()`, `-`
- Multiplication: `tf.multiply()`, `*`
- Division: `tf.divide()`, `/`  

```python
tensor=tf.constant([[1, 2], [3, 4]])
print("Original tensor:\n", tensor)
# basic operations
# Addition
added_tensor = tf.add(tensor, 10)
print("Added tensor:\n", added_tensor)
# Subtraction
subtracted_tensor = tf.subtract(tensor, 2)
print("Subtracted tensor:\n", subtracted_tensor)
# Multiplication
multiplied_tensor = tf.multiply(tensor, 3)
print("Multiplied tensor:\n", multiplied_tensor)
# Division
divided_tensor = tf.divide(tensor, 2)
print("Divided tensor:\n", divided_tensor)  
```

# performing basic arithmetic operations on tensors

```python
# performing basic arithmetic operations on tensors
tensor_a = tf.constant([[1, 2], [3, 4]])
tensor_b = tf.constant([[5, 6], [7, 8]])
sum_tensor = tf.add(tensor_a, tensor_b)
print("Sum of tensors:\n", sum_tensor)
diff_tensor = tf.subtract(tensor_b, tensor_a)
print("Difference of tensors:\n", diff_tensor)
prod_tensor = tf.multiply(tensor_a, tensor_b)
print("Product of tensors:\n", prod_tensor)
div_tensor = tf.divide(tensor_b, tensor_a)
print("Division of tensors:\n", div_tensor)
# matrix multiplication
matmul_tensor = tf.matmul(tensor_a, tensor_b)
print("Matrix multiplication of tensors:\n", matmul_tensor)
# using TensorFlow functions within a @tf.function for performance optimization
@tf.function
def compute_operations(x, y):
    return tf.add(x, y), tf.matmul(x, y)    
add_result, matmul_result = compute_operations(tensor_a, tensor_b)
print("Add result from tf.function:\n", add_result)     
print("Matmul result from tf.function:\n", matmul_result)
```

## aggregatiing tensors 

```python
# aggregatiing tensors = condensing a tensor down from multiple values to a smaller amount of values
tensor_3d = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("Original tensor:\n", tensor_3d)  
print("Sum of all elements in tensor:\n", tf.reduce_sum(tensor_3d).numpy())
print("Mean of all elements in tensor:\n", tf.reduce_mean(tensor_3d).numpy())       
print("Absolute value of -5:\n", tf.abs(-5).numpy())
print("Absolute values of tensor elements:\n", tf.abs(tensor_3d).numpy())
```
